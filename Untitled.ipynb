```python
# building a deep learning algorithm model
import tensorflow as tf
from tensorflow.keras import layers, callbacks
from tensorflow import keras
import numpy as np 
import pandas as pd
```

```python
clean_df=pd.read_csv("clean_df.csv")
```

```python
clean_df
```

```python
clean_df.head(7)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical features
categorical_features = ['Airline', 'Source', 'Destination', 'Route', 'Total_Stops', 'Additional_Info']

label_encoders = {}
for feature in categorical_features:
    label_encoders[feature] = LabelEncoder()
    clean_df[feature] = label_encoders[feature].fit_transform(clean_df[feature])
```

```python
def convert_duration(duration):
    # Split the duration into components
    duration = duration.split(' ')
    
    # Initialize hours and minutes
    hours = 0
    minutes = 0
    
    # Iterate over the split duration parts
    for part in duration:
        if 'h' in part:
            hours = int(part.replace('h', ''))
        elif 'm' in part:
            minutes = int(part.replace('m', ''))
    
    return hours * 60 + minutes

# Apply the function to the Duration column
clean_df['Duration'] = clean_df['Duration'].apply(convert_duration)
```

```python
# step one build network
#separate the target from the features
x = clean_df.copy()
y =x.pop("Price")
```

```python
# split the data into train and val set 
from sklearn.model_selection import train_test_split
x_train,x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
network = keras.Sequential([
    
    layers.Dense(units=100,  input_shape=[11]), # input layer
    layers.Dropout(0.02),
    layers.BatchNormalization(),
    layers.Activation("relu"),
    
    layers.Dense(units=360),
    layers.Dropout(0.02),
    layers.BatchNormalization(),
    layers.Activation("tanh"), # first hidden layers
    
    layers.Dense(units=512),# second hidden layers
    layers.Dropout(0.02),
    layers.BatchNormalization(),
    layers.Activation("sigmoid"),
    
    layers.Dense(units=512),# second hidden layers
    layers.Dropout(0.02),
    layers.BatchNormalization(),
    layers.Activation("relu"),
    
    
    layers.Dense(units=512),# second hidden layers
    layers.Dropout(0.02),
    layers.BatchNormalization(),
    layers.Activation("relu"),
    
    layers.Dense(units=1) #output
    
])
```

```python
# compile the model
network.compile(optimizer="adam", loss="mae"  , metrics=["mae"]    )
```

```python
# train the model
```

```python
stop= callbacks.EarlyStopping(min_delta=0.001, patience=10, restore_best_weights=True)
```

```python
hs = network.fit(x=x_train,
    y=y_train,
    batch_size=500,
    epochs=2000, validation_data=(x_val, y_val), callbacks=[stop])
```

```python
# setup plotting 
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
# set Matplotlib defaults
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',titleweight='bold', titlesize=18, titlepad=10)

import pandas as pd
pd.DataFrame(hs.history)[["loss", "val_loss"]].plot()
```
